Le dataset **Breast Cancer Wisconsin (Diagnostic)** contient 32 colonnes, dont la premi√®re est l'identifiant de l'√©chantillon, repr√©sentant le patient, et la derni√®re indique si la tumeur est b√©nigne ou maligne. Les autres colonnes d√©crivent les caract√©ristiques de la tumeur :

1. Rayon, p√©rim√®tre, aire : Mesures g√©om√©triques de la tumeur. \


2. Texture : Indique l'homog√©n√©it√© ou l'h√©t√©rog√©n√©it√© de la tumeur, bas√©e sur les niveaux de gris des images.  

3. Compacit√© : Rapport entre le p√©rim√®tre et la racine carr√©e de l'aire, refl√©tant la r√©gularit√© de la forme.

4. Sym√©trie : Mesure de la sym√©trie de la tumeur par rapport √† son centre.
5. Dimension fractale : Mesure de la complexit√© de la fronti√®re de la tumeur.

Ces mesures sont obtenues apr√®s une **biopsie** et une **analyse microscopique**, souvent √† l'aide d'un ***oculaire microm√©trique*** ou d'***un logiciel d'analyse d'image.***

**Comme les tumeurs ne sont ni parfaitement homog√®nes ni circulaires, plusieurs mesures sont effectu√©es sur diff√©rentes sections**.

 Pour chaque caract√©ristique, trois valeurs sont calcul√©es :

1. **Mean** : Moyenne des mesures, repr√©sentant la valeur g√©n√©rale de la caract√©ristique.
2. **SE** (Erreur standard) : Variabilit√© des mesures √† travers les sections.
3. **Worst** : Valeur maximale, indiquant le cas extr√™me.

Par exemple, pour le p√©rim√®tre :

**Mean** repr√©sente la taille moyenne de la tumeur. \

**SE** √©value la constance du p√©rim√®tre entre les sections. \

**Worst** refl√®te la taille maximale de la tumeur dans une section donn√©e.
Ces valeurs permettent une √©valuation plus pr√©cise des caract√©ristiques de la tumeur.

 Cela garantira que nous nous concentrons sur les caract√©ristiques les plus pertinentes pour la pr√©diction de notre cible.

Cette approche m√©thodique nous permettra d‚Äôoptimiser la performance de notre mod√®le tout en simplifiant sa structure ! ‚ú®

Analysons les paires de variables explicatives de la matrice de corr√©lation **(corr√©lation > 0,5)** afin d'identifier et d'**√©liminer** l'**une des variables fortement corr√©l√©es**. \

Cette d√©marche vise √† att√©nuer la multicolin√©arit√© qui, dans le contexte de notre probl√®me de classification, peut d√©grader significativement la performance du mod√®le. \

En effet, un tel probl√®me de multicolin√©arit√© pourrait engendrer une variance √©lev√©e des pr√©dictions, ce qui augmente le risque de sur-apprentissage et diminue la g√©n√©ralisation du mod√®le √† de nouvelles donn√©es. \

Cela se traduit par une baisse de la pr√©cision des classifications et une instabilit√© des r√©sultats.

```python
high_corr_var = np.where(corr> 0.5)
high_corr_var = [(corr.columns[x], corr.columns[y]) for x, y in zip(*high_corr_var) if x != y and x < y]

print(high_corr_var)
```

Maintenant, pour chaque couple de variables, on va essayer de d√©terminer quelle variable est la plus corr√©l√©e (*correlation biserial*) avec notre variable cible.

Note ‚ùó: La **corr√©lation bis√©riale** est une mesure de la relation entre une variable continue et une variable dichotomique (c'est-√†-dire une variable qui ne peut prendre que deux valeurs, Maligne/B√©gnine pour les tumeurs suspectes dans notres cas).

Elle est en fait un **cas particulier** de la *corr√©lation de Pearson*, adapt√©e √† des situations o√π l'une des variables est dichotomique.

La corr√©lation bis√©riale est donn√©e par la formule :

$$r_b = \frac{\overline{X}_1 - \overline{X}_0}{\sigma_X} \times \frac{p_1 p_0}{y}$$

O√π :
- $ ( \overline{X}_1 ) $ est la moyenne de la variable continue pour les observations de la cat√©gorie 1.
- $ ( \overline{X}_0 )$  est la moyenne de la variable continue pour les observations de la cat√©gorie 0.
- $ ( \sigma_X )$  est l'√©cart-type de la variable continue.
- $ ( p_1 )$  est la proportion d'observations dans la cat√©gorie 1.
- $ ( p_0 )$  est la proportion d'observations dans la cat√©gorie 0.
- $ ( y )$  est la hauteur de l'ordonn√©e de la courbe normale correspondant √† la proportion \( p_1 \).



```python
# on va initialiser un data frame pour contenir chaque variable et sa correlation avec la variable cible
corr_with_target_df = pd.DataFrame(columns=['variable', 'correlation_biserial_with_diagnosis'])



# It√©rer sur les couples de variables hautement corr√©l√©es
for var1, var2 in high_corr_var:
    # Calculer la corr√©lation bis√©riale
    corr, _ = pointbiserialr(X_train[var1], y_train)
    corr1, _1 = pointbiserialr(X_train[var2], y_train)
    new_row = pd.DataFrame({'variable': [var1], 'correlation_biserial_with_diagnosis': [corr]})
    new_row_1 = pd.DataFrame({'variable': [var2], 'correlation_biserial_with_diagnosis': [corr1]})
    # Ajouter un nouveau row au DataFrame de r√©sultats
    corr_with_target_df = pd.concat([corr_with_target_df, new_row], ignore_index=True)
    corr_with_target_df = pd.concat([corr_with_target_df, new_row_1], ignore_index=True)
# Afficher le DataFrame de r√©sultats
print(corr_with_target_df)
```

```python
corr_with_target_df.head()
```

Le code ci-dessous, calcule le coefficient de corr√©lation bis√©rial pour chaque variable parmi les paires les plus corr√©l√©es.
Il √©limine la variable la moins corr√©l√©e avec la variable cible ('Diagnosis') en comparant les coefficients de corr√©lation bis√©riale.
Apr√®s cette s√©lection, il supprime les redondances.

```python
# Cr√©er une liste vide pour stocker les indices des lignes √† conserver
indices_to_keep = []

# It√©rer sur le DataFrame en comparant les valeurs cons√©cutives
for i in range(len(corr_with_target_df) - 1):
    if corr_with_target_df['correlation_biserial_with_diagnosis'][i] > corr_with_target_df['correlation_biserial_with_diagnosis'][i+1]:
        indices_to_keep.append(i)
        #print("La variable la plus impactante sur la variable cible entre", corr_with_target_df.iloc[i, 0], "et", corr_with_target_df.iloc[i+1, 0], "est :", corr_with_target_df.iloc[i, 0])
    else:
        indices_to_keep.append(i+1)
        #print("La variable la plus impactante sur la variable cible entre", corr_with_target_df.iloc[i, 0], "et", corr_with_target_df.iloc[i+1, 0], "est :", corr_with_target_df.iloc[i+1, 0])

# Cr√©er un nouveau DataFrame contenant uniquement les lignes s√©lectionn√©es repr√©sentant les variables les plus impactantes sur la cible
top_correlated_features_with_target = corr_with_target_df.iloc[indices_to_keep, :]

# Supprimer les lignes r√©p√©t√©es (bas√© sur les noms des variables)
top_correlated_features_with_target_unique = top_correlated_features_with_target.drop_duplicates(subset='variable')
print(top_correlated_features_with_target_unique.head())
variables_names_to_keep_list = top_correlated_features_with_target_unique['variable'].tolist()
print("Les variables les plus impactantes sur la variable cible parmi la totalit√© des variables sont :", variables_names_to_keep_list)
```

Nous avons finalement obtenu les variables **sans redondance**, list√©es ci-dessous, qui sont les plus corr√©l√©es de mani√®re bis√©riale et donc **significatives pour notre variable cible**.               
Nous les conserverons tout au long de notre √©tude, dans le Data frame intitul√© : *X_train_feature_selected.*

```python
X_train_feature_selected = X_train[variables_names_to_keep_list]
X_train_feature_selected.head()
X_train_feature_selected.info()
```

1.   Si les donn√©es **suivent une loi normale** (ou gaussienne), la **standardisation** sera la m√©thode privil√©gi√©e.

2.   Si les donn√©es **ne** suivent **pas** une loi **normale**, il sera plus appropri√© d‚Äôutiliser **la normalisation (Min-Max scaling)**, qui ram√®ne les valeurs dans une plage d√©finie (g√©n√©ralement [0, 1]).

La normalisation des donn√©es, via le min-max scaling, doit se faire uniquement sur l'ensemble d'entra√Ænement (train) et ensuite √™tre appliqu√©e aux donn√©es de test (test) en utilisant les param√®tres calcul√©s sur l'ensemble d'entra√Ænement. \

Concr√®tement, il faut calculer les valeurs minimales et maximales √† partir des donn√©es d'entra√Ænement afin de mettre √† l'√©chelle les donn√©es entre 0 et 1 (ou un autre intervalle), en suivant cette formule : \


$X_{norm} = \frac{X - \min_{\text{train}}}{\max_{\text{train}} - \min_{\text{train}}}$


o√π $(min_{{train}})$ est la valeur minimale des donn√©es d'entra√Ænement, \

et $(max_{{train}})$ est la valeur maximale des donn√©es d'entra√Ænement.

**Puis, les m√™mes param√®tres (min et max) utilis√©s pour normaliser les donn√©es d'entra√Ænement, doivent √™tre appliqu√©s aux donn√©es de test.**

 Cela permet d'assurer une mise √† l'√©chelle coh√©rente et *pr√©vient la fuite de donn√©es*, garantissant ainsi un apprentissage efficace du mod√®le.

# Traitement des outliers par imputation des donn√©es
Pour traiter les outliers, nous rempla√ßons les valeurs extr√™mes par des valeurs plus repr√©sentatives, telles que la moyenne ou la m√©diane des donn√©es non outli√®res, pour chaque variable.

Cette m√©thode permet de r√©duire l'impact des outliers sur l'analyse tout en pr√©servant l'int√©grit√© des donn√©es restantes.

```python
# Fonction pour remplacer les valeurs aberrantes par la m√©diane
def replace_outliers_with_median(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    borne_inferieure = Q1 - 1.5 * IQR
    borne_superieure = Q3 + 1.5 * IQR

    # Identifier les valeurs aberrantes
    valeurs_aberrantes = data[(data[column] < borne_inferieure) | (data[column] > borne_superieure)]

    # Remplacer les valeurs aberrantes par la m√©diane des non-aberrantes
    if not valeurs_aberrantes.empty:
        mediane_non_aberrantes = data[(data[column] >= borne_inferieure) & (data[column] <= borne_superieure)][column].median()
        data.loc[valeurs_aberrantes.index, column] = mediane_non_aberrantes

    return data

# Remplacer les valeurs aberrantes pour chaque colonne
for column in X_train_selected_scaled.columns:
    X_train_selected_scaled_no_outliers = replace_outliers_with_median(X_train_selected_scaled, column)
```

Nous allons √† nouveau utiliser les boxplots üìä pour √©valuer les am√©liorations obtenues apr√®s le traitement des outliers.

 Ces visualisations nous permettront de comparer la distribution des donn√©es avant et apr√®s imputation, en mettant en √©vidence la r√©duction significative des valeurs extr√™mes ‚ö†Ô∏è.

 Gr√¢ce √† cette approche, nous pouvons mieux appr√©cier la qualit√© des donn√©es et la robustesse de notre mod√®le.

En d'autres termes, les algorithmes de classification (surtout Random Forest, qu'on va d'ailleurs inclure dans notre √©tude ) **ont tendance √† favoriser la classe majoritaire (sains ou "1") car elle est plus pr√©sente dans les donn√©es**. \
Cela peut conduire √† un mod√®le qui pr√©dit souvent "sain" m√™me quand l'individu est en r√©alit√© malade, ce qui g√©n√®re des faux n√©gatifs. \

‚ùó Un **taux √©lev√© de faux n√©gatifs** pose **un probl√®me majeur**, notamment dans une √©tude m√©dicale.

**Dans notre cas, cel√† signifie que des patients atteints de cancer du sein pourraient √™tre incorrectement diagnostiqu√©s comme sains**.

En cons√©quence, ces patients ne recevraient pas le traitement n√©cessaire, ce qui pourrait mettre leur vie en danger ‚ö†Ô∏è.

# R√©√©quilibrage des donn√©es avec SMOTE (Synthetic Minority Over-sampling Technique)

Pour traiter les probl√®mes de d√©s√©quilibre des classes, une approche couramment utilis√©e est le **SMOTE (Synthetic Minority Over-sampling Technique)**.

**SMOTE** est une technique de sur√©chantillonnage qui vise √† √©quilibrer les classes en cr√©ant de nouveaux exemples synth√©tiques √† partir des instances de la classe minoritaire ( qui la classe de tumeurs malignes (M) dans notre cas ).

Gr√¢ce √† **SMOTE**, nos mod√®les de machine learning peuvent mieux apprendre les caract√©ristiques de la classe minoritaire (M), ce qui contribue √† r√©duire les erreurs comme les faux n√©gatifs, particuli√®rement critiques dans notre contexte comme mentionn√© pr√©cedemment.

```python
# Application du SMOTE
smote = SMOTE(random_state=42)
X_train_pretreated, y_train_pretreated= smote.fit_resample(X_train_selected_scaled_no_outliers, y_train)
```

# R√©gression Logistique (avant l'optimisation des hyperparam√®tres) üîçüìä

Entra√Ænons le mod√®le de r√©gression logistique et mesurons les m√©triques en utilisant l'ensemble des variables. \
 formule de la r√©gression logistique est :
$P(Y=1|X) = \frac{1}{1 + e^{-(\beta_0 + \beta_1X_1 + \beta_2X_2 + \ldots + \beta_nX_n)}}$
o√π: P(Y=1|X) repr√©sente la probabilit√© que la variable d√©pendante Y soit √©gale √† 1, et $$\beta_0$$, $$\beta_1$$,$$\ldots$$,$$\beta_n$$ sont les coefficients estim√©s.




```python
# Cr√©er un mod√®le de r√©gression logistique
model = LogisticRegression()

# Entra√Æner le mod√®le sur les donn√©es d'entra√Ænement
model.fit(X_train_pretreated, y_train_pretreated)

# Faire des pr√©dictions sur les donn√©es de test
y_pred = model.predict(X_test_selected_scaled_no_outliers)
y_proba = model.predict_proba(X_test_selected_scaled_no_outliers)[:, 1]  # Probabilit√©s pour la classe positive

# Calculer les m√©triques
accuracy_no_opt = accuracy_score(y_test, y_pred)
precision_no_opt = precision_score(y_test, y_pred)
recall_no_opt = recall_score(y_test, y_pred)
f1_no_opt = f1_score(y_test, y_pred)
confusion_mat_no_opt = confusion_matrix(y_test, y_pred)

# Cr√©er un tableau de r√©sultats
results_logistic_no_opt = pd.DataFrame({
    'M√©trique': ['Accuracy', 'Pr√©cision', 'Rappel', 'F1-Score'],
    'Valeur': [accuracy_no_opt, precision_no_opt, recall_no_opt, f1_no_opt]
})

# Afficher le tableau de r√©sultats
print(results_logistic_no_opt)

# Afficher la matrice de confusion
print("Matrice de confusion:")
print(confusion_mat_no_opt)

# Heatmap de la matrice de confusion
cm = confusion_mat_no_opt
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix Heatmap")
plt.show()
```

## R√©gression Logistique (apr√®s l'optimisation des hyperparam√®tres) üîçüìä

Pour optimiser notre mod√®le de r√©gression logistique, nous allons recourir √† **une Grid Search**. Dans ce cas pr√©cis, l'utilisation d'une Random Search n'est pas indispensable, car nous n'avons que trois hyperparam√®tres √† consid√©rer :

**Penalty** ‚öñÔ∏è : Une fonction utilis√©e pour ajouter une p√©nalit√© aux coefficients du mod√®le afin de pr√©venir le surapprentissage (overfitting). Les types courants incluent l1 (r√©gularisation Lasso) et l2 (r√©gularisation Ridge).

**C **üìè : Un hyperparam√®tre de r√©gularisation qui contr√¥le la force de la p√©nalit√©. Une valeur √©lev√©e de C indique moins de r√©gularisation, tandis qu'une valeur basse augmente la r√©gularisation.

**Solver** üõ†Ô∏è : L'algorithme utilis√© pour optimiser les coefficients du mod√®le. Les solveurs courants incluent liblinear (adapt√© aux petits ensembles de donn√©es) et saga (capable de g√©rer des ensembles de donn√©es plus grands et d'impl√©menter √† la fois l1 et l2).


La Grid Search est appropri√©e ici √©tant donn√© le nombre limit√© d'hyperparam√®tres. En revanche, une Random Search est plus b√©n√©fique dans des sc√©narios o√π le nombre d'hyperparam√®tres est √©lev√©, comme c'est souvent le cas avec les for√™ts al√©atoires.
 En effet, tester toutes les combinaisons possibles dans des mod√®les complexes peut non seulement n√©cessiter une grande capacit√© de m√©moire, mais aussi entra√Æner des temps de calcul significatifs.

```python
# Cr√©er un mod√®le de r√©gression logistique
model = LogisticRegression()

# D√©finir les distributions de param√®tres pour RandomizedSearchCV
param_dist = {
    'penalty': ['l1', 'l2'],
    'C': np.logspace(-4, 4, 20),
    'solver': ['liblinear', 'saga']
}

# Cr√©er un objet RandomizedSearchCV
random_search = RandomizedSearchCV(model, param_distributions=param_dist, n_iter=10, cv=5)

# Entra√Æner le mod√®le sur les donn√©es d'entra√Ænement
random_search.fit(X_train_pretreated, y_train_pretreated)

# Obtenir le meilleur mod√®le
best_model = random_search.best_estimator_


# Pr√©dictions sur les donn√©es d'entra√Ænement et de test
y_pred_train = best_model.predict(X_train_pretreated)
y_pred_test = best_model.predict(X_test_selected_scaled_no_outliers)

# Calculer et imprimer les m√©triques d'√©valuation
accuracy_train_logistic = accuracy_score(y_train_pretreated, y_pred_train)
accuracy_test_logistic = accuracy_score(y_test, y_pred_test)

precision_logistic_final = precision_score(y_test, y_pred_test)
recall_logistic_final = recall_score(y_test, y_pred_test)
f1_logistic_final = f1_score(y_test, y_pred_test)
conf_matrix_logistic_final = confusion_matrix(y_test, y_pred_test)

results_logistic_final = pd.DataFrame({
    'M√©trique': ['Accuracy (Train)', 'Accuracy (Test)', 'Pr√©cision', 'Rappel', 'F1-Score'],
    'Valeur': [accuracy_train_logistic, accuracy_test_logistic, precision_logistic_final, recall_logistic_final, f1_logistic_final]
})

print(results_logistic_final)
print("\nMatrice de confusion:")
print(conf_matrix_logistic_final)

# Heatmap de la matrice de confusion
sns.heatmap(conf_matrix_logistic_final, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Pr√©dit")
plt.ylabel("R√©el")
plt.title("Heatmap de la Matrice de Confusion")
plt.show()

```

Les **performances** üìà du mod√®le de r√©gression logistique ont √©t√© **significativement am√©lior√©es** üöÄ gr√¢ce √† **l'hyperoptimisation des hyperparam√®tres**, sans overfitting ( vu que la train_accuracy est presque √©gale √† test_accuracy...) ! \

De plus, le taux de Faux N√©gatifs a diminu√© de 50 % ‚ùó\

N.B ‚ùó : *Cette am√©lioration est particuli√®rement importante car **un faux n√©gatif** peut avoir des cons√©quences graves pour les patients, entra√Ænant **un retard dans le traitement et une aggravation de leur √©tat de sant√©**.\
 En am√©liorant le Recall, nous augmentons notre capacit√© √† sauver des vies et √† fournir un diagnostic plus fiable.*

Visualisons les scores de chaque m√©trique avant et apr√®s l'optimisation via la Grid Search üîç dans le **barplot ci-dessous**. \
Cela nous permettra de mieux appr√©cier l'am√©lioration obtenue üéâ.


```python
results_logistic_final = results_logistic_final.set_index('M√©trique')['Valeur'].to_dict()
results_logistic_final = {
    'Accuracy': results_logistic_final['Accuracy (Test)'],
    'Precision': results_logistic_final['Pr√©cision'],
    'Recall': results_logistic_final['Rappel'],
    'F1-Score': results_logistic_final['F1-Score']
}

results_logistic_no_opt = results_logistic_no_opt.set_index('M√©trique')['Valeur'].to_dict()
results_logistic_no_opt = {
    'Accuracy': results_logistic_no_opt['Accuracy'],
    'Precision': results_logistic_no_opt['Pr√©cision'],
    'Recall': results_logistic_no_opt['Rappel'],
    'F1-Score': results_logistic_no_opt['F1-Score']
}

# D√©finir les m√©triques √† tracer
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']

# Obtenir les valeurs pour chaque m√©trique avant et apr√®s l'optimisation
before_values = [results_logistic_no_opt[metric] for metric in metrics]
after_values = [results_logistic_final[metric] for metric in metrics]

# D√©finir la largeur des barres et les positions
bar_width = 0.35
index = np.arange(len(metrics))

# Cr√©er le graphique √† barres
fig, ax = plt.subplots()
rects1 = ax.bar(index, before_values, bar_width, label='Avant Optimisation', color='skyblue')
rects2 = ax.bar(index + bar_width, after_values, bar_width, label='Apr√®s Optimisation', color='coral')

# D√©finir les √©tiquettes, le titre et la l√©gende
ax.set_xlabel('M√©trique')
ax.set_ylabel('Valeur')
ax.set_title('Comparaison des M√©triques Avant et Apr√®s Optimisation')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(metrics, rotation=45, ha='right')
ax.legend()

# Afficher le graphique
plt.tight_layout()
plt.show()

```

La Random Forest est un algorithme d'**apprentissage supervis√©** qui construit plusieurs arbres de d√©cision pour am√©liorer la pr√©cision des pr√©dictions.
Il combine les r√©sultats de ces arbres par un vote majoritaire (pour la classification) ou par une moyenne (pour la r√©gression), r√©duisant ainsi le risque de surapprentissage.

La pr√©diction $(\hat{y})$ est exprim√©e par :

o√π $(N)$ est le nombre d'arbres et $(f_i(x))$ est la pr√©diction de l'arbre $(i)$ pour l'entr√©e $(x)$.

```python
# D√©finir la plage de valeurs de k √† tester
k_values = list(range(1, 31))

# Initialiser la liste pour stocker les taux d'erreur
error_rates = []

# Boucle √† travers les valeurs de k et calculer le taux d'erreur
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_pretreated, y_train_pretreated)
    y_pred = knn.predict(X_test_selected_scaled_no_outliers)
    error_rate = 1 - accuracy_score(y_test, y_pred)  # Taux d'erreur = 1 - pr√©cision
    error_rates.append(error_rate)

# Trouver la valeur de k avec le plus bas taux d'erreur
min_error_rate = min(error_rates)
best_k = k_values[error_rates.index(min_error_rate)]

# Tracer le taux d'erreur en fonction de la valeur de k
plt.figure(figsize=(10, 6))
plt.plot(k_values, error_rates, marker='o', linestyle='-', label='Taux d\'Erreur')

# Mettre en √©vidence la meilleure valeur de k
plt.scatter(best_k, min_error_rate, color='red', s=100, zorder=5, label=f'Meilleur k = {best_k}')
plt.axvline(x=best_k, linestyle='--', color='gray')  # Projection sur l'axe x

# Ajouter une annotation textuelle pour montrer la meilleure valeur de k et le taux d'erreur
plt.text(best_k, min_error_rate + 0.01, f'k={best_k}\nErreur={min_error_rate:.4f}',
         ha='center', color='red')

# Titre et √©tiquettes du trac√©
plt.title('Taux d\'Erreur vs. Valeur de K pour kNN')
plt.xlabel('Valeur de K')
plt.ylabel('Taux d\'Erreur')
plt.grid(True)
plt.legend()
plt.show()
```

D'apr√®s le graphique lin√©aire ci-dessus repr√©sentant le taux d'erreur en fonction de la valeur de k, k=7 est la meilleure option, avec l'error rate le plus bas ( error rate = 2.9 % ).\

Maintenant, visualisons comment l'algorithme SVM d√©termine l'hyperplan en utilisant les deux premi√®res composantes principales pour la visualisation.

```python
# Appliquer la PCA pour la r√©duction de dimension
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_train_pretreated)

# Cr√©er un mod√®le SVM
svm_model = SVC(kernel='linear')

# Entra√Æner le mod√®le
svm_model.fit(X_pca, y_train_pretreated)

# Obtenir l'hyperplan de s√©paration
w = svm_model.coef_[0]
a = -w[0] / w[2]
b = -w[1] / w[2]
xx, yy = np.meshgrid(np.linspace(-2, 4, 30), np.linspace(-2, 4, 30))  # Augmenter la r√©solution du meshgrid
zz = a * xx + b * yy - (svm_model.intercept_[0]) / w[2]

# Tracer les points de donn√©es et l'hyperplan
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Tracer les points de donn√©es avec des couleurs plus claires
scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y_train_pretreated, cmap='coolwarm', s=50, edgecolors='k')

# Ajouter une barre de couleur pour distinguer les deux classes
colorbar = plt.colorbar(scatter, ax=ax, pad=0.1)
colorbar.set_label('Classes')

# Tracer l'hyperplan avec une couleur plus distinctive
ax.plot_surface(xx, yy, zz, color='green', alpha=0.5)

# Ajuster l'angle de vue pour une meilleure visualisation de la s√©paration
ax.view_init(elev=20, azim=120)

# √âtiquettes et titres
ax.set_xlabel("Composante Principale 1")
ax.set_ylabel("Composante Principale 2")
ax.set_zlabel("Composante Principale 3")
ax.set_title("Visualisation de l'Hyperplan SVM avec S√©paration Am√©lior√©e")

# Afficher le trac√©
plt.show()
```
